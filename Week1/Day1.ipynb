# Day 1 – Introduction to Quantum Computing

## What is Quantum Computing?

Classical computers work using **bits**, which can be:
- 0 or 1

Quantum computers use **qubits**, which can be:
- 0  
- 1  
- or both at the same time  

This ability increases computing power and helps solve complex problems.

---

## Classical vs Quantum

- Classical Bit → 0 or 1  
- Qubit → 0 or 1 or both simultaneously  

---

## Why Quantum Computing is Needed?

In real-world problems:
- There are too many possibilities
- Classical computers check one by one
- This makes the process slow

Quantum computers:
- Process multiple possibilities at once
- Solve complex problems faster

---

## Real World Applications

- Healthcare → Drug Discovery  
- Chemistry → Molecular Simulation  
- Finance → Portfolio Optimization  
- Artificial Intelligence → Complex Model Training  
- Cybersecurity → Cryptography

## Superposition

Superposition is a phenomenon where:
- A qubit can exist in multiple states at the same time

Example:

| Qubits | Possible States |
|--------|---------------|
| 1      | 2             |
| 2      | 4             |
| 3      | 8             |
| 10     | 1024          |

👉 More qubits = exponentially more possibilities

---

## Entanglement

- Two qubits become connected
- Changing one affects the other instantly

👉 Acts like a single system

---

## Interference

- Amplifies correct answers  
- Suppresses incorrect answers  

👉 Helps in getting the desired output

---

## Decoherence

The biggest challenge in quantum computing:

- Quantum information is fragile  
- Noise can destroy computations  

Process:
- Ideal State  
↓  
- Environmental Noise  
↓  
- Information Loss  
↓  
- Decoherence  

---

## Limitations of Quantum Computing

- Not suitable for everyday tasks  
- Very sensitive to noise  
- Still under development  

---

## Why Quantum Computing Exists

As problems grow:
- More variables  
- More combinations  
- More computations  

👉 Classical systems struggle  

👉 Quantum computers handle this complexity better
